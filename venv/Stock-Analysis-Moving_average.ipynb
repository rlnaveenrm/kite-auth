```python
# processing
# remmove the weekend values from the dataframe
sbin_data=sbin_data[(sbin_data['Close'])!=0]
sbin_data["3weekMA"]=sbin_data["Close"].rolling(15).mean()
sbin_data
sbin_data['Close'].plot()
sbin_data['3weekMA'].plot()
#plt.plot(sbin_data['Close'].values.tolist()[7:-7])
#plt.plot(sbin_data['3weekMA'].values.tolist()[14:])

```
